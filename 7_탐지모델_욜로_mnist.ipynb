```python
import pandas as pd
from ultralytics import YOLO

# --------------------------------------------------------------------------
# ## 1. 10개 데이터셋으로 학습
# --------------------------------------------------------------------------
print("--- 10개 데이터셋으로 학습을 시작합니다 ---")

# 새 모델 로드
model_small= YOLO('yolov8n.pt')
data_model_small = './datasets/mnist/data.yaml' 
model_small.train(data=data_model_small,
                  epochs=30,
                  imgsz=640,
                  project='runs/detect',
                  name='small_mnist')
```

```python
# ## 2. 62개데이터셋으로 학습
# --------------------------------------------------------------------------
print("\n--- 62개데이터셋으로 학습을 시작합니다 ---")

# 새 모델 다시 로드
model_large = YOLO('yolov8n.pt')
data_model_large= './datasets/mnist_2/data.yaml'
model_large.train(data=data_model_large ,
                  epochs=30,
                  imgsz=640,
                  project='runs/detect',
                  name='large_mnist'
                  )
```

```python
import pandas as pd

# --------------------------------------------------------------------------
# ## 3. 두 모델 성능 비교
# --------------------------------------------------------------------------

def compare_model_performance(small_model_path, large_model_path):
    """
    두 YOLO 모델의 훈련 결과(results.csv)를 비교하여 성능 지표를 출력합니다.

    Args:
        small_model_path (str): 데이터가 적은 모델의 results.csv 파일 경로.
        large_model_path (str): 데이터가 많은 모델의 results.csv 파일 경로.
    """
    try:
        # CSV 파일 로드 및 컬럼명 공백 제거
        df_small = pd.read_csv(small_model_path).rename(columns=lambda c: c.strip())
        df_large = pd.read_csv(large_model_path).rename(columns=lambda c: c.strip())

        # 비교할 핵심 성능 지표
        metrics = ['metrics/mAP50(B)', 'metrics/mAP50-95(B)']

        # 각 모델의 마지막 epoch 성능 추출하여 비교 데이터프레임 생성
        comparison_df = pd.DataFrame({
            'small_mnist (10개)': df_small.iloc[-1][metrics],
            'large_mnist (62개)': df_large.iloc[-1][metrics]
        })

        # 인덱스 이름 변경
        comparison_df.index = ['mAP@.50', 'mAP@.50-.95']

        # 결과 출력
        print("--- 두 모델의 성능 비교를 시작합니다 ---")
        print("\n[ 최종 Epoch 성능 비교표 ]")
        print(comparison_df)

    except FileNotFoundError as e:
        print(f"오류: 파일을 찾을 수 없습니다. 경로를 확인해주세요.\n{e}")
    except Exception as e:
        print(f"처리 중 오류가 발생했습니다: {e}")

# --- 실행 ---
# 각 모델의 훈련 결과 파일 경로
path_small = 'runs/detect/small_mnist/results.csv'
path_large = 'runs/detect/large_mnist/results.csv'

# 함수 호출
compare_model_performance(path_small, path_large)
```

```python
import pandas as pd

# --------------------------------------------------------------------------
# ## 3. 세 가지 모델 성능 텍스트로 출력
# --------------------------------------------------------------------------
print("--- 세 가지 모델의 성능을 텍스트로 출력합니다 ---")

# 각 모델의 훈련 결과 파일 경로
path_small = 'runs/detect/small_mnist/results.csv'
path_large = 'runs/detect/large_mnist/results.csv'
path_final = 'runs/detect/large_mnist_final/results.csv' # 튜닝된 모델 경로

try:
    # CSV 파일들을 pandas DataFrame으로 읽어오기
    df_small = pd.read_csv(path_small).rename(columns=lambda c: c.strip())
    df_large = pd.read_csv(path_large).rename(columns=lambda c: c.strip())
    df_final = pd.read_csv(path_final).rename(columns=lambda c: c.strip())

    # 각 모델의 마지막 epoch 성능 추출
    scores_small = df_small.iloc[-1]
    scores_large = df_large.iloc[-1]
    scores_final = df_final.iloc[-1]
    
    # 텍스트 출력
    print("\n### 1. small_mnist (10개 데이터)")
    print(f"* mAP@.50: {scores_small['metrics/mAP50(B)']:.5f}")
    print(f"* mAP@.50-.95: {scores_small['metrics/mAP50-95(B)']:.5f}")
    
    print("\n### 2. large_mnist (62개 데이터)")
    print(f"* mAP@.50: {scores_large['metrics/mAP50(B)']:.5f}")
    print(f"* mAP@.50-.95: {scores_large['metrics/mAP50-95(B)']:.5f}")

    print("\n### 3. large_mnist_final (62개 데이터 + 튜닝)")
    print(f"* mAP@.50: {scores_final['metrics/mAP50(B)']:.5f}")
    print(f"* mAP@.50-.95: {scores_final['metrics/mAP50-95(B)']:.5f}")

except FileNotFoundError as e:
    print(f"\n오류: 파일을 찾을 수 없습니다. 경로를 확인해주세요.")
    print(f"누락된 파일: {e.filename}")
except Exception as e:
    print(f"\n처리 중 오류가 발생했습니다: {e}")
```

## 왜 튜닝 후 성능 변화 없음?
이는 오류가 아니라, 현재 모델의 상태를 알려주는 중요한 단서입니다. 몇 가지 이유를 추측해 볼 수 있습니다.

데이터셋의 한계 (가장 유력한 원인)
62장이라는 데이터셋은 여전히 매우 작은 규모입니다. 모델이 데이터로부터 학습할 수 있는 정보의 총량이 이미 한계에 도달했을 가능성이 높습니다. 즉, 공부법(하이퍼파라미터)을 아무리 바꿔도 교재(데이터)에 없는 내용은 더 배울 수 없는 것과 같습니다. 모델의 성능이 파라미터가 아닌 데이터의 양에 의해 발목 잡혀있는 '데이터 제한적(Data-Limited)' 상태인 것입니다.

YOLOv8의 뛰어난 기본 파라미터
Ultralytics의 YOLOv8 모델은 이미 수많은 실험을 통해 매우 잘 최적화된 기본(default) 하이퍼파라미터를 가지고 있습니다. MNIST처럼 비교적 단순한 데이터셋에서는 이 기본값이 이미 최적에 가까워, tune을 해도 더 나은 조합을 찾지 못했을 수 있습니다.

## 결론 
이번 실험의 가장 중요한 교훈은 다음과 같습니다.

현재 모델의 성능을 한 단계 더 끌어올리기 위한 방법은 하이퍼파라미터 튜닝이 아니라, '더 많은 데이터'를 확보하는 것입니다.

성능이 더 이상 오르지 않는 병목 현상이 나타났을 때, 그 원인이 파라미터인지 데이터인지를 파악하는 것은 매우 중요합니다. 이번 결과는 그 원인이 명백히 데이터 부족에 있음을 보여준다.
